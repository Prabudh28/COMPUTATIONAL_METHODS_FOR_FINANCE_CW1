## 2. Data Handling 

### Part (a) Download Historical Data for Two Equities

### Selecting two equities from FTSE 100 with a two-year period using yfinance:

#### Importing Necessary Libraries

```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#### Defining stock tickers(Unilever: ULVR and HSBC: HSBA.L) and date range:2022-2024

```python
tickers = ['ULVR.L', 'HSBA.L']  # Unilever and HSBC on the London Stock Exchange
start_date = '2022-01-01'
end_date = '2024-01-01'

# Download data
data = yf.download(tickers, start=start_date, end=end_date)['Adj Close']
data.columns = ['Unilever', 'HSBC']  # Rename columns for convenience
data.dropna(inplace=True)  # Remove any missing data
print(data.head())  # Display first few rows of data
```

### Part (b) Plotting Price Movements and Calculating Annualized Returns and Standard Deviations

```python
# Plotting the price movements of Unilever and HSBC
plt.figure(figsize=(12, 6))
plt.plot(data['Unilever'], label="Unilever (ULVR.L)", color="blue")
plt.plot(data['HSBC'], label="HSBC (HSBA.L)", color="green")
plt.title("Stock Price Movements (Jan 2022 - Jan 2024)")
plt.xlabel("Date")
plt.ylabel("Price (GBP)")
plt.legend()
plt.show()
```

#### Calculating daily returns based on the data

```python
daily_returns = data.pct_change().dropna()  # Percentage change to calculate daily returns
print(daily_returns.head())  # Display first few rows of daily returns
```

#### Calculating annualized average returns

```python
annualized_avg_returns = (1 + daily_returns.mean()) ** 252 - 1
print("Annualized Average Returns:")
print(annualized_avg_returns)
```

#### Calculating annualised standard deviations

```python
annualized_std_devs = daily_returns.std() * np.sqrt(252)
print("Annualized Standard Deviations:")
print(annualized_std_devs)
```

#### After analysing Unilever and HSBC over the past two years, we observe distinct differences in their growth rates and volatility.
#### Annualised Average Returns: Unilever and HSBC show differing annualised returns, reflecting their growth potential. Higher returns indicate stronger performance, making a stock more attractive to growth-focused investors.
#### Annualised Standard Deviation (Volatility): Volatility, as measured by annualised standard deviation, indicates risk. A higher standard deviation suggests more price fluctuation, signaling higher risk. Stocks with lower volatility may appeal to risk-averse investors, while higher-volatility stocks may attract those with higher risk tolerance. 
#### Risk-Return Balance: Our findings align with the risk-return trade-off principle, where higher returns often come with higher risk.
#### Investment Insights: Comparing the two stocks highlights diversification potential; a balanced portfolio can benefit from including stocks with varying risk and return profiles. The analysis also reflects how market factors affect each stock differently, with HSBC potentially more sensitive to economic shifts and Unilever showing more stability due to steady consumer demand.
